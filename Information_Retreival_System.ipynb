```python
import wikipedia
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
```

```python
def system():
  a="ab"
  articles=[]
  while (a!="q"):
    inp=input("Enter Article to be downloaded >")
    articles.append(inp)
    a=input("Press q to quit or any other key to continue >")
  query=input("Enter the query to be searched against >")
  wiki_lst=[]
  title=[]
  for article in articles:
    wiki_lst.append(wikipedia.page(article).content)
  tfidf=TfidfVectorizer()
  train_tfidf = tfidf.fit_transform(wiki_lst) 
  query_vec = tfidf.transform([query])
  sim1=cosine_similarity(train_tfidf,query_vec)
  data={"Article":wiki_lst,"Similarity Score":sim1.flatten()}
  df=pd.DataFrame(data)
  df.sort_values(["Similarity Score"],ascending=False, inplace=True)
  len=int(input("Enter the number of relevant results to be displayed"))
  sc=input("Do you want the esimilarity score as well ? [y/n] ")
  if sc=="y":
    print(display(df.head(len).style.hide_index()))
  else:
    for i in len(range(len)):
      print(f"ARTICLE {i+1}:\n")
      print(df[i])
      print("\n")   
```

```python
system()
```
